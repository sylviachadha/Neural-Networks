# Implementation Of Perceptron

```python
import tensorflow as tf

# Define input features and labels
x = tf.constant([[0.4, 0.3], [0.2, 0.7], [0.8, 0.1]])
y = tf.constant([[1], [1], [0]])

# Define weights and bias
w = tf.Variable(tf.random.normal([2, 1]))
b = tf.Variable(tf.random.normal([1]))

# Define perceptron model
def perceptron(x):
    output = tf.matmul(x, w) + b     # matmul or matrix multiplication used 
    return tf.nn.sigmoid(output)

# Define loss function and optimizer
loss_fn = tf.keras.losses.BinaryCrossentropy()
optimizer = tf.keras.optimizers.SGD(learning_rate=0.1)

# Train the model
for i in range(100):
    with tf.GradientTape() as tape:
        # Make predictions using the perceptron model
        y_pred = perceptron(x)
        # Calculate loss
        loss = loss_fn(y, y_pred)
    # Calculate gradients
    gradients = tape.gradient(loss, [w, b])
    # Update weights and bias using gradients
    optimizer.apply_gradients(zip(gradients, [w, b]))
    
# Print final weights and bias
print("Final weights:", w.numpy())
print("Final bias:", b.numpy())
```
