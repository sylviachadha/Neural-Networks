## Perceptron - for binary classification only.

- Most basic form of neural n/w which is in form of peceptron, u can consider it to be a single cell neuron in which there is only 1 node,neuron for information processing in between, apart from Input and Output nodes.
- For information processing within that single neuron, we have 2 functions, adder function & activation function.
- Then we have error function & finally we have an optimizer which helps us in reducing error we have obtained in a neural n/w model.
- **There can be as many i/p's for perceptron but it is valid for only binary output.**

## Multi-layer Perceptron

For MLP there is a library in Scikit learn called as MLPClassifier. Multi-layer perceptron is made up of a number of single perceptrons.

#### from sklearn.neural_network import MLPClassifier

Basic structures of MLP can be prepared in Scikit learn.

## Multilayer Perceptron

- MLP is a collection of a number of perceptrons.
- Whatever information is received at P1 (shown below) by using adder & activation function will act as input for the next layer.
- Below the input layer and 1st hidden layer make up total 3 perceptrons -

- Now this 1st hidden layer is input for next layer. So we have a 3 perceptrons each with 3 inputs and single o/p's as shown.
- Now in this 1st hidden layer there are 3 i/p nodes, these 3 will provide some information which will be added together & which will be activated.
- This continues till we reach last o/p neuron.
- In last perceptron, q1,q2,q3 will be input and y' will be the perceptron.

- The procedure remains same in perceptron and MLP. Once you get o/p, u compare it with the true o/p/target o/p and perform backpropagation.
- Same 2 phases. Forward propagation/phase where information is processed and backpropagation/phase where error is minimized and new weights r generated.


## Multilayer Perceptron - A recall of forward pass and backward propagation


- When we use scikit learn library, initial weights and bias are automatically generated.
- Below there will be many samples and 2 features x1 and x2, each sample will be passed 1 by 1.
- x1 and x2 are given features and h1 is the information which we have obtained (i.e. numeric value of h1 will be generated) with help of perceptron model.
- As shown below each node of pervious layer is connected with each node of next layer.
- For input layer to 1st hidden layer we have 2 * 3 matrix, 2 rows and 3 columns.
- For hidden layer to output layer we have 3 * 1 matrix, 3 rows and 1 column.

### <font color=purple> *The complexity of a network is determined by the total number of weights that are there in the model*

- This is because we know in every step information is being processed with help of connection b/w the two neurons & that is being done with help of weights that r provided to it.
    
- In below diagram we have 6 weights for hidden layer and 3 weights for o/p layer & hence total 9 weights. If also add biases then 9 (6+3 biases) for hidden and 4 (3+1 bias) for o/p layer, so total 13 parameters.
- Except for input layer i.e. for input features there are no additonal bias, all other layers (including hidden and output layer) will have bias attached to every neuron in that layer.
- In the LHS of figure, we see there are 3 perceptrons (x1,x2,h1), (x1,x2,h2), (x1,x2,h3). In RHS of figure we see there is 1 perceptron (h1,h2,h3,y0). In a perceptron there can be any number of inputs but only 1 output.
- f(H1), f(H2), f(H3) are the values we get after applying activation function.


### How to decide layers as well number of neurons in each layer?

- The number of hidden layers as well neurons in each layer, will depend on data you r having that is the number of features u r having, the number of samples u r having.
- In MLP's generally upto 2 to 3 layers is good enough for moderate data size when data is not very complex.
- Increasing hidden layers will lead to increase in complexity of model. The more model is complex, the more time it will take, and may even result in similar type of accuracy.
- In MLP's more things to fine tune compared to perceptron.

- Only the calculation part will increase from layers point of view otherwise the method will remain the same in case of backpropagation also.
- The number of layers we add means we r accumulating more error.
- Backpropagation means we have to reduce the error that was generated in each step. The error which we generated in each step will also be minimized for each step.
- Steepest gradient descent does not behave very well when there r several number of layers because it has to go back, we have to see the rate of change, the derivative sometimes it becomes a bit complex mathematically.

# <font color=blue>3 MLPClassifier for binary Classification


    The multilayer perceptron (MLP) is a feedforward artificial neural network model that maps input data sets to a set of appropriate outputs. 
    An MLP consists of multiple layers and each layer is fully connected to the following layer. 
    The nodes of the layers are neurons with nonlinear activation functions, except for the nodes of the input layer. 
    Between the input and the output layer there may be one or more nonlinear hidden layers.

Neural n/w's can be used for both regression and classification problems but many of ML algorithms r more suited for classification problems & similar for NN, better suited for classification problems.

### 3.2 Data pre-processing

# 3.3 MLPClassifier
### Example of an MLP having 2 hidden layers. The first layer has 5 neurons and the second layer has 2 neurons
### After defining the number of hidden layers, we need to specify the maximum number of iterations, the activation function and the optimizer (solver)

The network structure created in the process would look like:


Like MLPClassifier we also have MLP Regressor.

https://scikit-learn.org/stable/modules/generated/sklearn.neural_network.MLPRegressor.html

Whether we r solving a classification or regression problem, the layers,neurons i.e. architectures we have to define, the changes will be how we r calculating error & how we r going further.

```python
# You need to define a tuple for hidden layer. Length of tuple tells number of 
# hidden layers and values tell number of neurons in each layer. eg hidden_layer_sizes=(5,2)

from sklearn.neural_network import MLPClassifier
mlp_clf = MLPClassifier(hidden_layer_sizes=(5,2),
                        max_iter = 300, activation = 'relu',
                        solver = 'adam') ## solver stands for optimizer ##
```

- **hidden_layer_sizes** : With this parameter we can specify the number of layers and the number of nodes we want to have in the Neural Network Classifier. Each element in the tuple represents the number of nodes at the ith position, where i is the index of the tuple. Thus, the length of the tuple indicates the total number of hidden layers in the neural network.

- **max_iter:** Indicates the number of epochs.

- **activation:** The activation function for the hidden layers.

- **solver:** This parameter specifies the algorithm for weight optimization over the nodes.

All of above are hyperparameters.e.g. you can try different optimization algorithms e.g for optimizer sgd can also be used which is faster in case you have smaller n/w as sgd works on gradients, so sometimes convergence can be faster in that.

#### Practically, we need to have higher number of neurons and higher number of hidden layers. 
##### So building an MLP, with the following specifications:
    
    No of hidden layers - 3
    first layer - 150 neurons
    second layer - 100 neurons
    third layer - 50 neurons
    Activation function - relu
    Solver - adam

1. One is the manual way i.e initially guess how many layers & neurons to keep & then do hit & trial with different values.
2. Other is systematic way i.e. Grid search CV in which we can do fine tuning for various parameters.

Initially we make some guess to get some initial idea to see what is my accuracy and what is confusion matrix. Then based on that we can fine tune network further.

```python
mlp_clf = MLPClassifier(hidden_layer_sizes=(150,100,50),
                        max_iter = 300,activation = 'relu',
                        solver = 'adam')

mlp_clf.fit(trainX_scaled, trainY)
```

# 5 Final notes

#### GridSearchCV is a technique to search through the best parameter values from the given set of the grid of parameters. 
    It is basically a cross-validation (CV) method. 
    The model and the parameters are required to be fed in. 
    Best parameter values are extracted and then the predictions are made.
    Primarily, it takes 4 arguments i.e. estimator, param_grid, cv, and scoring. 

        estimator – A scikit-learn model
        param_grid – A dictionary with parameter names as keys and lists of parameter values.
        scoring – The performance measure. For example, ‘r2’ for regression models, ‘precision’ for classification models.
        cv – An integer that is the number of folds for K-fold cross-validation.
        
#### K fold cross validation is a kind of **<font color=blue>regularization method** when u have a limited number of dataset/samples, when dataset is not very big. How u r taking samples of given data.

The general technique for spotting overfitting is known as cross validation. Cross validation involves reserving samples of the training data on which you can later test your model to check that it is able to generalise well. **There could be different types of CV like Hold-out cross validation, K-fold cross validation, Leave-one-out cross validation.**




    

#### **<font color=blue>Data Preprocessing:** In data preprocessing, Check for nans, separate the data into input and output, and also scale the data. Standard Scalar: for scaling the data. Besides this u can also check correlations among the input features you have. This way u can reduce the size of features you have. 
    
#### <font color=blue>Performance metrics    
    TN / True Negative: when a case was negative and predicted negative
    TP / True Positive: when a case was positive and predicted positive
    FN / False Negative: when a case was positive but predicted negative
    FP / False Positive: when a case was negative but predicted positive

#### <font color=green>Precision – What percent of your predictions were correct?

    Precision is the ability of a classifier not to label an instance positive that is actually negative. For each class it is defined as the ratio of true positives to the sum of true and false positives.

    TP – True Positives
    FP – False Positives

    **Precision – Accuracy of positive predictions.**
    Precision = TP/(TP + FP)
    
#### <font color=green>Recall – What percent of the positive cases did you catch? 

    Recall is the ability of a classifier to find all positive instances. For each class it is defined as the ratio of true positives to the sum of true positives and false negatives.

    FN – False Negatives

    Recall: Fraction of positives that were correctly identified.
    Recall = TP/(TP+FN)
    
#### F1 score – What percent of positive predictions were correct? 

    The F1 score is a weighted harmonic mean of precision and recall such that the best score is 1.0 and the worst is 0.0.    
    Generally speaking, F1 scores are lower than accuracy measures as they embed precision and recall into their computation. 
    As a rule of thumb, the weighted average of F1 should be used to compare classifier models, not global accuracy.

    F1 Score = 2*(Recall * Precision) / (Recall + Precision)

## Some Extra Notes

1. **Bias-Variance trade in ML**

Bias-Variance trade off will be applicable in context of ANN same like as in ML. Whenever we have dataset and some optimizations applied we will have some interest bias and inherent bias-variance trade off also. There is a tradeoff between a model’s ability to minimize bias and variance.

- Linear Regression is often a high bias low variance ml model if we call LR as a not complex model. Simpler models are generally High Bias, Low Variance models.
- **<font color=blue>Bias** is the difference between the average prediction of our model and the correct value which we are trying to predict.Model with high bias pays very little attention to the training data and oversimplifies the model. It always leads to high error on training and test data.
- **<font color=blue>Variance** is the variability of model prediction for a given data point or a value which tells us spread of our data. Model with high variance pays a lot of attention to training data and does not generalize on the data which it hasn’t seen before. As a result, such models perform very well on training data but has high error rates on test data.



Lasso Regression are like regularization methods that help you in maintaining the bias and the variance that may be there in the model.It is a result of overfitting or underfitting when sample is not proper;y distributed. 

The purpose of all regularization techniques (Like L1,L2,K-fold) is to minimize the bias and the variance. It is to reduce the effects of overfitting.
